# DATAGENERATOR EXAMPLE
Please, note that a DataGenerator is not meant to be used as it is in this example. A DataGenerator is meant to be used to feed Keras training, prediction and evaluation. This example is only meant to provide a better understanding on how the generator works.

### Import the DataGenerator and auxiliary libraries

```python
from datagenerator import DataGenerator
import matplotlib.pyplot as plt
```

### Create the DataGenerator
The first two parameters are the SSS data and GT paths. Within these paths, only files TRAN09.png and TRAN09.png will be used. Also, patches will be provided in random order. The margin, step and batch size are set to 41, 41 and 10 by default.

```python
theGenerator=DataGenerator('DATASET/DATA','DATASET/GT',transectList=[8,9],doRandomize=True)
```

### Print some basic info

```python
print(' * NUMBER OF BATCHES: '+str(theGenerator.__len__()))
print(' * BATCH SIZE: '+str(theGenerator.batchSize))
X,y=theGenerator.__getitem__(theGenerator.__len__()-1)
print(' * LAST BATCH SIZE : '+str(X.shape[0]))
print(' * DATA SHAPE: ('+str(X.shape[1])+','+str(X.shape[2])+')')
print(' * NUMBER OF CLASSES: '+str(y.shape[3]))
```

### Plot the last batch
First row in the SSS data and second row in the ground truth.

```python
plt.figure()
for i in range(X.shape[0]):
    plt.subplot(2,X.shape[0],i+1)
    plt.imshow(X[i,:,:,0],cmap='gray',vmin=0,vmax=1)
    plt.axis('off')
    plt.subplot(2,X.shape[0],X.shape[0]+i+1)
    plt.imshow(y[i])
    plt.axis('off')
plt.show()
```
